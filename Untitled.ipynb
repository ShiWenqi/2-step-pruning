```python
import json
import numpy as np
from openpyxl import Workbook
from openpyxl.chart import (
    LineChart,
    BarChart,
    Reference,
    Series,
)

with open('./log_original.json') as fp:
    original_data = json.load(fp)
with open('./log_prune.json') as fp:
    prune_data = json.load(fp)
with open('./log_prune_layer.json') as fp:
    prune_layer_data = json.load(fp)

bandwidth0 = np.power(32, 2) * 3

# Ratio of Computing resources between mobile device and cloud server
# Power of  835 ~ 3W
# Power of  E5-2640 ~95W
R = 10

# Transmission Rate
Rate = 200


data_len = 0
for data in original_data:
    data_len = data_len+1
    
###############################
# original
###############################

layer_cfg = ['', '']
acc = ['acc_original', '']
# delta_t = ['time_original', 0]
delta_ts = 0
time_remain = 0
# delta_t_computations = ['', '']
bandwidth = ['bandwidth_original', bandwidth0]
transmission_delay = ['transmission_delay_original', bandwidth0/Rate]
# all_conv_computations = ['', '']
for data in original_data:
    layer_cfg.append(data['layer_cfg'])
    acc.append(data['acc'])
    # delta_t.append(data['delta_t'])
    delta_ts += np.array(data['delta_ts'])
    # delta_t_computations.append(data['delta_t_computations'])
    bandwidth.append(data['bandwidth'])
    transmission_delay.append(data['bandwidth']/Rate)
    # all_conv_computations.append(data['all_conv_computations'])
    
time_original = ['time_original', 0]+list(delta_ts / len(original_data)) 
time_original_device = ['Device_computation', 0]+list(delta_ts / len(original_data)*R) 
time_remain = ['Cloud_computation', delta_ts[-1]/ len(original_data)]+list((delta_ts[-1]-delta_ts) / len(original_data))

wb = Workbook()
# ws = wb.create_sheet("original", 0)
ws = wb.active
ws.append(layer_cfg)
ws.append(acc)
# ws.append(delta_t)
ws.append(time_original)
ws.append(time_original_device)
ws.append(time_remain)
# ws.append(delta_t_computations)
ws.append(bandwidth)
ws.append(transmission_delay)
# ws.append(all_conv_computations)

###############################
# prune
###############################

layer_cfg = ['', '']
acc = ['acc_prune_s1', '']
# delta_t = ['time_prune', 0]
delta_ts = 0
# delta_t_computations = ['', '']
bandwidth = ['bandwidth_prune_s1', bandwidth0]
transmission_delay = ['Transmission', bandwidth0/Rate]
# all_conv_computations = ['', '']
for data in prune_data:
    layer_cfg.append(data['layer_cfg'])
    acc.append(data['acc'])
    # delta_t.append(data['delta_t'])
    delta_ts += np.array(data['delta_ts'])
    # delta_t_computations.append(data['delta_t_computations'])
    bandwidth.append(data['bandwidth'])
    transmission_delay.append(data['bandwidth']/Rate)
    # all_conv_computations.append(data['all_conv_computations'])
    
time_original = ['time_prune_s1', 0]+list(delta_ts / len(original_data))
time_original_device = ['Device_computation', 0]+list(delta_ts / len(original_data)*R)
time_remain = ['Cloud_computation', delta_ts[-1]/ len(original_data)]+list((delta_ts[-1]-delta_ts) / len(original_data))

# ws = wb.create_sheet("prune_layer", 0)
for i in range(3):
    ws.append(list())
ws.append(layer_cfg)
ws.append(acc)
# ws.append(delta_t)
ws.append(time_original)
ws.append(time_original_device)
ws.append(time_remain)
# ws.append(delta_t_computations)
ws.append(bandwidth)
ws.append(transmission_delay)
# ws.append(all_conv_computations)

###############################
# prune_layer
###############################

layer_cfg = ['', '']
acc = ['acc_prune_s2', '']
# delta_t = ['time_prune_layer', 0]
delta_ts = 0
# delta_t_computations = ['', '']
bandwidth = ['bandwidth_prune_s2', bandwidth0]
transmission_delay = ['Transmission', bandwidth0/Rate]
# all_conv_computations = ['', '']
for data in prune_layer_data:
    layer_cfg.append(data['layer_cfg'])
    acc.append(data['acc'])
    # delta_t.append(data['delta_t'])
    delta_ts += np.array(data['delta_ts'])
    # delta_t_computations.append(data['delta_t_computations'])
    bandwidth.append(data['bandwidth'])
    transmission_delay.append(data['bandwidth']/Rate)
    # all_conv_computations.append(data['all_conv_computations'])

    
time_original = ['time_prune_s2', 0]+list(delta_ts / len(original_data))
time_original_device = ['Device_computation', 0]+list(delta_ts / len(original_data)*R)
time_remain = ['Cloud_computation', delta_ts[-1]/ len(original_data)]+list((delta_ts[-1]-delta_ts) / len(original_data))

# ws = wb.create_sheet("prune_layer", 0)
for i in range(3):
    ws.append(list())
ws.append(layer_cfg)
ws.append(acc)
# ws.append(delta_t)
ws.append(time_original)
ws.append(time_original_device)
ws.append(time_remain)
# ws.append(delta_t_computations)
ws.append(bandwidth)
ws.append(transmission_delay)
# ws.append(all_conv_computations)

for i in range(3):
    ws.append(list())
for index in range(len(layer_cfg)):
    if isinstance(layer_cfg[index], int):
        layer_cfg[index] = 'conv'
    elif layer_cfg[index] == '' and index == 1:
        layer_cfg[index] = 'og image'
    elif layer_cfg[index] == '':
        pass
    elif layer_cfg[index] == 'L':
        layer_cfg[index] ='Avg Pool'
    else:
        layer_cfg[index] = 'Max Pool'

ws.append(layer_cfg)


# draw chart
time_original = Reference(ws, min_col=1, min_row=3, max_col=data_len)
time_original_device = Reference(ws, min_col=1, min_row=4, max_col=data_len)
time_remain_original = Reference(ws, min_col=1, min_row=5, max_col=data_len)

time_prune = Reference(ws, min_col=1, min_row=13, max_col=data_len)
time_prune_device = Reference(ws, min_col=1, min_row=14, max_col=data_len)
time_remain_prune = Reference(ws, min_col=1, min_row=15, max_col=data_len)

time_prune_layer = Reference(ws, min_col=1, min_row=23, max_col=data_len)
time_prune_layer_device = Reference(ws, min_col=1, min_row=24, max_col=data_len)
time_remain_prune_layer = Reference(ws, min_col=1, min_row=25, max_col=data_len)

bandwidth_original = Reference(ws, min_col=1, min_row=6, max_col=data_len)
bandwidth_prune = Reference(ws, min_col=1, min_row=16, max_col=data_len)
bandwidth_prune_layer = Reference(ws, min_col=1, min_row=26, max_col=data_len)

transmission_delay_original = Reference(ws, min_col=1, min_row=7, max_col=data_len)
transmission_delay_prune = Reference(ws, min_col=1, min_row=17, max_col=data_len)
transmission_delay_prune_layer = Reference(ws, min_col=1, min_row=27, max_col=data_len)

# line chart for time
c1 = LineChart()
c1.add_data(time_original, titles_from_data=True, from_rows=True)
c1.add_data(time_prune, titles_from_data=True, from_rows=True)
c1.add_data(time_prune_layer, titles_from_data=True, from_rows=True)

cats = Reference(ws, min_col=2, min_row=31, max_col=data_len)
c1.set_categories(cats)

c1.x_axis.title = 'layers'
c1.y_axis.title = 'time elapsed (ms)'
c1.y_axis.majorGridlines = None
c1.title = 'bandwidth/time'

# bar chart for bandwidth
c2 = BarChart()
c2.add_data(bandwidth_original, titles_from_data=True, from_rows=True)
c2.add_data(bandwidth_prune, titles_from_data=True, from_rows=True)
c2.add_data(bandwidth_prune_layer, titles_from_data=True, from_rows=True)
c2.y_axis.axId = 200  # set axid other than 100
c2.y_axis.title = 'data volume (B)'

# Display y-axis of the second chart on the right by setting it to cross the x-axis at its maximum
c1.y_axis.crosses = "max"
c1 += c2

ws.add_chart(c1, "D4")

c3 = BarChart()
c3.add_data(transmission_delay_original, titles_from_data=True, from_rows=True)
c3.add_data(time_original_device, titles_from_data=True, from_rows=True)
c3.add_data(time_remain_original, titles_from_data=True, from_rows=True)
c3.y_axis.title = 'Lantancy (ms)'
c3.x_axis.title = 'layers'
c3.type = "col"
c3.overlap = 100
c3.grouping = "stacked"
c3.set_categories(cats)
ws.add_chart(c3, "D5")

c4 = BarChart()
c4.add_data(transmission_delay_prune, titles_from_data=True, from_rows=True)
c4.add_data(time_prune_device, titles_from_data=True, from_rows=True)
c4.add_data(time_remain_prune, titles_from_data=True, from_rows=True)
c4.y_axis.title = 'Lantancy (ms)'
c4.x_axis.title = 'layers'
c4.type = "col"
c4.overlap = 100
c4.grouping = "stacked"
c4.set_categories(cats)
ws.add_chart(c4, "D5")


c5 = BarChart()
c5.add_data(transmission_delay_prune_layer, titles_from_data=True, from_rows=True)
c5.add_data(time_prune_layer_device, titles_from_data=True, from_rows=True)
c5.add_data(time_remain_prune_layer, titles_from_data=True, from_rows=True)
c5.y_axis.title = 'Lantancy (ms)'
c5.x_axis.title = 'layers'
c5.type = "col"
c5.overlap = 100
c5.grouping = "stacked"
c5.set_categories(cats)
ws.add_chart(c5, "D6")


wb.save('chart.xlsx')
```
